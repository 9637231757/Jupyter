```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset

df = pd.read_csv('Amazon_Sales_data.csv')
```

```python
# first 5 rows of the dataset

df.head()
```

```python
# shape of the dataset

df.shape
```

```python
# type of data types of each column

df.dtypes
```

```python
# Convert the 'Order Date' column to datetime format

df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# printing output

print(df['Order Date'])
```

```python
# Extract month and year from 'Order Date' column

df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
df['Year_Month'] = df['Order Date'].dt.strftime('%Y-%m')

print(df)
```

```python
# updated dataset

df.head()
```

```python
# Check the unique values in 'Month' column

df['Month'].unique()
```

```python
# Check the unique values in 'Year' column

df['Year'].unique()
```

```python
# Check the unique values in 'Year_Month' column

df['Year_Month'].unique()
```

```python
# Group the data by 'Month' and calculate the total sales for each month

monthly_sales = df.groupby('Month')['Units Sold'].sum().reset_index()

monthly_sales
```

#### plot monthly Sales trend

```python
# Plot the monthly sales trend

plt.figure(figsize=(10,6))
sns.lineplot(x='Month', y='Units Sold', data=monthly_sales)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.show()
```

```python
# Group the data by 'Year' and calculate the total sales for each year

yearly_sales = df.groupby('Year')['Units Sold'].sum().reset_index()

yearly_sales
```

#### Plot a yearly sales trend

```python

# Plot the yearly sales trend

plt.figure(figsize=(10,6))
sns.lineplot(x='Year', y='Units Sold', data=yearly_sales)
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Units Sold')
plt.show()
```

```python
# Group the data by 'Year_Month' and calculate the total sales for each year-month

year_month_sales = df.groupby('Year_Month')['Units Sold'].sum().reset_index()

year_month_sales
```

#### Plot the yearly-monthly sales trend

```python
# Plot the yearly-monthly sales trend

plt.figure(figsize=(10,6))
sns.lineplot(x='Year_Month', y='Units Sold', data=year_month_sales)
plt.title('Year-Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

#### Top 10 Items with highest sales

```python
# Find the top 10 products with highest sales
top_products = df.groupby('Item Type')['Units Sold'].sum().reset_index().sort_values(by='Units Sold', ascending=False).head(10)

top_products
```

```python
# Plot the top 10 products with highest sales

plt.figure(figsize=(10,6))
sns.barplot(x='Item Type', y='Units Sold', data=top_products)
plt.title('Top 10 Products with Highest Sales')
plt.xlabel('Item Sold')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

#### Top 10 Countries with highest sales

```python
# Find the top 10 states with highest sales

top_Countries = df.groupby('Country')['Units Sold'].sum().reset_index().sort_values(by='Units Sold', ascending=False).head(10)

top_Countries
```

#### Calculating Units sold for each item

```python
# Calculate total units sold for each item type

item_type_units_sold = df.groupby('Item Type')['Units Sold'].sum().reset_index()
```

```python

# Plot the total units sold for each item type

plt.figure(figsize=(10,6))
plt.bar(item_type_units_sold['Item Type'], item_type_units_sold['Units Sold'])
plt.xlabel('Item Type')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold by Item Type')
plt.show()
```

#### Calculate unit sold for each channel

```python
# Calculate total units sold for each sales channel

sales_channel_units_sold = df.groupby('Sales Channel')['Units Sold'].sum().reset_index()
```

```python
# Plot the total units sold for each sales channel

plt.figure(figsize=(10,6))
plt.bar(sales_channel_units_sold['Sales Channel'], sales_channel_units_sold['Units Sold'])
plt.xlabel('Sales Channel')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold by Sales Channel')
plt.show()
```

#### Calculation total units sold for each priority

```python

# Calculate total units sold for each order priority

order_priority_units_sold = df.groupby('Order Priority')['Units Sold'].sum().reset_index()
```

```python

# Plot the total units sold for each order priority
plt.figure(figsize=(10,6))
plt.bar(order_priority_units_sold['Order Priority'], order_priority_units_sold['Units Sold'])
plt.xlabel('Order Priority')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold by Order')
plt.show()
```
